# 3. Hipoteza

Ta hipoteza se nanaša na spol tekmovalcev oz. njihovo uspešnost. Moja hipoteza je, da je moški spol bolj uspešen. Bomo pa analizirali podatke tudi za moške.

```python
Tekmovalke = podatki[podatki.spol == "ženska"]
Število_tekmovalk_za_določeno_leto = Tekmovalke.groupby("leto").size()
Število_tekmovalk_za_določeno_leto.plot.bar(x = "leto",figsize = (15,7))
```

```python
Tekmovalci = podatki[podatki.spol == "moški"]
Število_tekmovalcev_za_določeno_leto = Tekmovalci.groupby("leto").size()
Število_tekmovalcev_za_določeno_leto.plot.bar(x = "leto", figsize = (15,7))
```

Iz teh dveh grafov vidimo, da čeprav je tekmovalcev precej več, pa število tekmovalk z leti hitreje narašča. Pogeljmo še uspešnost obeh spolov.

```python
Spol_tekmovalcev_in_njihov_dosežek = podatki_brez_HM[["leto", "spol", "nagrada"]].dropna()
Tekmovalke_in_njihov_dosežek = Spol_tekmovalcev_in_njihov_dosežek[Spol_tekmovalcev_in_njihov_dosežek.spol == "ženska"]
Število_medalij_tekmovalk_po_letih = Tekmovalke_in_njihov_dosežek.groupby("leto").size().sort_values(ascending=True)
Število_medalij_tekmovalk_po_letih.plot.bar(x = "leto", figsize = (15,7))
```

```python
Tekmovalci_in_njihov_dosežek = Spol_tekmovalcev_in_njihov_dosežek[Spol_tekmovalcev_in_njihov_dosežek.spol == "moški"]
Število_medalij_tekmovalcev_po_letih = Tekmovalci_in_njihov_dosežek.groupby("leto").size().sort_values(ascending=True)
Število_medalij_tekmovalcev_po_letih.plot.bar(x = "leto", figsize = (15,7))
```

```python
a = Število_tekmovalk_za_določeno_leto.rename("št. tekmovalk")
b = Število_medalij_tekmovalk_po_letih.sort_values(ascending=True).rename("št. medalj")
primerjava_število_medalij_s_številom_tekmovalk = pandas.merge(a,b, left_index=True, right_index=True)
#primerjava_število_medalij_s_številom_tekmovalk.head(25)
razmerje = primerjava_število_medalij_s_številom_tekmovalk["št. medalj"].div(primerjava_število_medalij_s_številom_tekmovalk["št. tekmovalk"])
razmerje.plot.bar(x="leto", figsize = (15,7))
```

Iz tega grafa vidimo, da se število medalj, ki jih tekmovalke dobijo počasi zmanšuje glede na število tekmovalk, kljub temu, pa zadnja leta približno tretino tekmovalk
doseže medaljo. Če pogledamo še za moški spol.

```python
a = Število_tekmovalcev_za_določeno_leto.rename("št. tekmovalcev")
b = Število_medalij_tekmovalcev_po_letih.sort_values(ascending=True).rename("št. medalj")
primerjava_število_medalij_s_številom_tekmovalcev = pandas.merge(a,b, left_index=True, right_index=True)
#primerjava_število_medalij_s_številom_tekmovalcev.head(25)
razmerje = primerjava_število_medalij_s_številom_tekmovalcev["št. medalj"].div(primerjava_število_medalij_s_številom_tekmovalcev["št. tekmovalcev"])
razmerje.plot.bar(x="leto", figsize = (15,7))
```

Iz stolpčnega grafa za moške tekmovalce, vidimo, da razmerje prav tako počasi pada, vendar pa zadnja leta(skoraj vsa leta), vsak drugi  tekmovalec dobi medaljo.
Naša hipoteza se je iskazala za pravilno(za natančnejšo analizo bi analizirali še ločeno po medaljah, ki jih določen spol doseže, vendar na podlagi teh podatkov, bi verjetno moški spol prevladoval, v vsaj dveh kategorijah)
